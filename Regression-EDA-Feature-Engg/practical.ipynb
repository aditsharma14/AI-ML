```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Data Cleaning

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset["Region"] = 0
split_idx = dataset[dataset["day"].astype(str).str.contains("Sidi-Bel Abbes", na=False)].index[0]
dataset.loc[split_idx:, "Region"] = 1
```

```python
dataset = dataset[dataset["day"] != "day"]      # drop the repeated header row from the second region block
dataset = dataset.dropna().reset_index(drop=True)  # drop the region-title row and the one genuinely malformed row
df = dataset
```

```python
df.info()
```

```python
df[["Region"]]=df[['Region']].astype(int)
```

```python
df.isnull().sum()
```

```python
df=df.dropna().reset_index(drop=True)
```

```python
df.columns=df.columns.str.strip()
df.columns
```

Changes the required columns as integer data type

```python
df.columns
```

```python
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

Changing the other columns to float data datatype

```python
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```
