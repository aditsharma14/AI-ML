```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
## create the dataset
X,y=make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=42)

```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
##Model Training
from sklearn.linear_model import LogisticRegression
LogisticRegressionModel=LogisticRegression()
```

```python
LogisticRegressionModel.fit(X_train,y_train)
```

```python
y_pred=LogisticRegressionModel.predict(X_test)
print(y_pred)
```

```python
LogisticRegressionModel.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
score=accuracy_score(y_test,y_pred)
print(score)
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

HyperParamter tuning and cross validation

```python
model=LogisticRegression()
penalty=['l1','l2','elasticnet']
c_values=[100,10,1,0.1,0.01]
solver=['newton-cg','lbfgs','liblinear','sag','saga']
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver)
```

```python
##GridSearchCV
from sklearn.model_selection import GridSearchCV,StratifiedKFold
CV=StratifiedKFold()
grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=CV,
                  n_jobs=-1)
```

```python
grid
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
grid.predict(X_test)
```

Randomized Search CV

```python
from sklearn.model_selection import RandomizedSearchCV
randomcv=RandomizedSearchCV(estimator=model,param_distributions=params,
                            cv=5,scoring='accuracy')
```

```python
randomcv
```

```python
randomcv.fit(X_train,y_train)
```

```python
randomcv.best_score_
```

```python
randomcv.predict(X_test)
```

```python
randomcv.predict_proba(X_test)
```
